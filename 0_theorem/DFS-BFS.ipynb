## 재귀

```python
def recursive_function(i):
    if i == 5:
        return None
    
    print(f"{i}번째 재귀함수에서 {i+1}번째 재귀함수를 호출한다.")
    recursive_function(i+1)
    print(f"{i}번째 재귀 함수를 종료한다.")
    
recursive_function(1)
```

## DFS
- 스택 혹은 재귀
- 시간 복잡도: $O(N)$

```python
def dfs(graph, v, visited):
    visited[v] = True
    print(v)
    # print(stack)
    
    for i in graph[v]:
        if not visited[i]:
            dfs(graph, i, visited)
            

graph = [
    [],
    [2, 3, 8],
    [1, 7],
    [1, 4, 5],
    [3, 5],
    [3, 4],
    [7],
    [2, 6, 8],
    [1, 7]
]

visited = [False] * 9

dfs(graph, 1, visited)
```

## BFS

```python
from collections import deque

def bfs(graph, start, visited):
    queue = deque([start])
    visited[start] = True
    
    while queue:
        v = queue.popleft()
        print(v)
        
        for i in graph[v]:
            if not visited[i]:
                queue.append(i)
                visited[i] = True
                
                
graph = [
    [],
    [2, 3, 8],
    [1, 7],
    [1, 4, 5],
    [3, 5],
    [3, 4],
    [7],
    [2, 6, 8],
    [1, 7]
]

visited = [False] * 9
bfs(graph, 1, visited)
```

## DFS vs BFS
- 시간복잡도는 두 개 모두 인접 리스트인 경우, $O(V+E)$, 인접 행렬인 경우 $O(V^2)$
- 구현상의 차이점은 `need_visited` 목록을 관리하는 데 있어서,
    - DFS는 스택을 사용하고,
    - BFS는 재귀를 사용한다!
- 실용상의 차이점
    - DFS는 미로찾기, 경로찾기
    - BFS는 최단 경로 찾기 등

```python
def dfs(graph, start_node):
    to_visit, visited = list(), list()
    to_visit.append(start_node)
    
    while to_visit:
        node = to_visit.pop()
        
        if node not in visited:
            visited.append(node)
            to_visit.extend(graph[node])
            
    return visited
```
