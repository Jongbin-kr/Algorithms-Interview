어우 자꾸 시간초과 나네....

```python
# import sys; input = lambda: sys.stdin.readline().rstrip()
N, M = map(int, input().split())

graph = []
for i in range(N):
    row = list(input())
    if "I" in row: j = row.index("I"); start_y, start_x = i, j
    graph.append(row)
print(*graph, sep='\n')

def dfs(start_y, start_x):
    to_visit = []
    visited = [[0] * M for _ in range(N)]
    meet_people = 0
    
    to_visit.append((start_y, start_x))
    
    while to_visit:
        y, x = to_visit.pop()
        
        for dy, dx in [(+1, 0), (-1, 0), (0, +1), (0, -1)]:
            if 0 <= y+dy < N and 0 <= x+dx < M and visited[y+dy][x+dx] == 0 and graph[y+dy][x+dx] != "X" and (y+dy, x+dx) not in to_visit:
                    to_visit.append((y+dy, x+dx))
                    visited[y][x] = 1
                    if graph[y][x] == "P": 
                        meet_people += 1
                    
    if meet_people: return meet_people
    else: return "TT"
    
print(dfs(start_y, start_x))
```
