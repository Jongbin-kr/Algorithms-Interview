```python
# import sys; input = lambda: sys.stdin.readline().rstrip()
from itertools import product
N = int(input())
paper = [list(map(int, input().split())) for _ in range(N)]


cnt_0 = 0
cnt_1 = 0
def divide_paper(paper, N):
    global cnt_0, cnt_1

    if len(paper) == 1:    ## 1*1의 종이로 나눠진 경우, 해당 원소를 확인 후 더하기
        if sum(*paper) == 0: cnt_0 += 1
        else: cnt_1 += 1
        return None
    
    flattened_sum = sum(sum(paper, []))
    if flattened_sum == 0:  ## 맨 처음부터 하나의 수로만 이뤄진 경우
        cnt_0 += 1
        return None
    
    elif flattened_sum == N**2:
        cnt_1 += 1
        return None
    
    else:
        M = int(N/2)
        for y_range, x_range in product([(0, M), (M, N)], repeat=2):
            start_y, end_y = y_range
            start_x, end_x = x_range
            part = list(map(lambda x: x[start_x:end_x], paper[start_y:end_y]))
            flattened_sum = sum(sum(part, []))
            
            if flattened_sum == M**2:
                cnt_1 += 1
            elif flattened_sum == 0:
                cnt_0 += 1
                
            else:
                divide_paper(part, M)

divide_paper(paper, N)

print(cnt_0, cnt_1, sep='\n')
```
