```python
from typing import *

n, m, y, x, k_orders = map(int, input().split())    #!! 원래 지문에서는 x가 N, 즉 세로 방향이고 y가 M, 즉 가로 방향이지만, 의미상 혼동을 피하고자 둘을 바꿨음

world = []
for _ in range(n):
    world.append(list(map(int, input().split())))
    
orders = map(int, input().split())
order_map = [(), (0, +1, '->'), (0, -1, '<-'), (-1, 0, '^'), (+1, 0, 'v')]


class Dice():
    def __init__(self, y: int, x: int):
        self.side = {'top': 0, 'bottom': 0, 'left': 0, 'right': 0, 'front': 0, 'back': 0}
        self.y = y
        self.x = x
        
        
    def roll(self, order: int):
        if order == 1:  # 오른쪽으로 굴리기
            self.side['bottom'], self.side['right'], self.side['top'], self.side['left'] = \
            self.side['right'], self.side['top'], self.side['left'], self.side['bottom']
            
        elif order == 2: # 왼쪽으로 굴리기
            self.side['bottom'], self.side['left'], self.side['top'], self.side['right'] = \
            self.side['left'], self.side['top'], self.side['right'], self.side['bottom']
            
        elif order == 3: # 위로 굴리기
            self.side['bottom'], self.side['back'], self.side['top'], self.side['front'] = \
            self.side['back'], self.side['top'], self.side['front'], self.side['bottom']
   
        else: # 아래로 굴리기
            self.side['bottom'], self.side['front'], self.side['top'], self.side['back'] = \
            self.side['front'], self.side['top'], self.side['back'], self.side['bottom']
                        
        dy, dx, _= order_map[order]
        self.y += dy
        self.x += dx


    def copy_or_cut(self):
        y, x = self.y, self.x
        if world[y][x]:     # 굴려서 도착한 칸에 쓰인 수가 0이 아닌 경우
            self.side['bottom'] = world[y][x]   # 해당 칸의 숫자를 주사위 바닥면으로 복사하고
            world[y][x] = 0                     # 칸에 쓰여있는 수는 0이 된다
            
        else:               # 굴려서 도착한 칸에 쓰인 수가 0인 경우
            world[y][x] = self.side['bottom']  # 주사위 바닥의 수가 칸으로 복사된다.



def check_within_map(order: int, dice: Dice) -> bool:
    y, x = dice.y, dice.x
    dy, dx, _ = order_map[order]
    
    if 0<= y+dy < n and 0<= x+dx < m:
        return True
    
    else:
        return False
            
            

dice = Dice(y, x)            
for order in orders:
    if check_within_map(order, dice):
        dice.roll(order)
        dice.copy_or_cut()
        
        print(dice.side['top'])
    
    else:
        continue
```
