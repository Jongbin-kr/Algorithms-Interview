증가하는 흐름이 끊어지는 경우, 어디로 돌아갈 것인가? 그냥 일일이 따져봐야하나?

```python
N = int(input())
ns = [0] + list(map(int, list(input().split())))

dp = [0] * (N+1)
dp[0] = 0
new_dp = [0] * (N+1)

for i in range(1, N):
    if ns[i] > ns[i-1]:
        if new_dp[i-1] != 0:
            new_dp[i] = new_dp[i-1] + 1
        
        dp[i] = max(new_dp[i])
    
    else:
        # find new max start value
        prev = i-1
        while ns[i] < ns[prev]:
            prev -= 1
        new_dp[i] = dp[prev] + 1
        
        dp[i] = dp[i-1]
```

# brute forece로 풀려고 한 것들

답안 참고해서 다시 풀이

아래처럼 하면 중간에 큰 수가 들어왔다가 이어지는 걸 할 수 없음.

```python
N = int(input())
nums = list(map(int, input().split()))

dp = [1] * N
for start in range(N-1):
    ns = nums[start:]
    prev_max = ns[0]
    cnt = 1
    for j in range(1, len(ns)):
        if ns[j] > prev_max:
            cnt += 1
            prev_max = ns[j]
            
    dp[start] = cnt
    
print(max(dp))
        
```

```python
N = int(input())
nums = list(map(int, input().split()))

dp = [1] * N
for i in range(1, N):
    for j in range(i):
        if nums[i] > nums[j]:
            dp[i] = max(dp[i], dp[j] + 1)
            
print(max(dp))
```
