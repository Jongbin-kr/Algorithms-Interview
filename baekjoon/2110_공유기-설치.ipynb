매개변수 탐색을 통해, 결정 문제를 최적화문제로 바꿔 풀어보자!

매개 변수: 공유기간 최대 거리
결정 함수: 해당 최대 거리로 공유기를 설치하면, c_wifi개의 공유기를 설치할 수 있는가?

```python
"""
완전 탐색을 통한 매개변수 탐색 -> 타임아웃!
"""
# import sys; input = lambda: sys.stdin.readline().rstrip()
n_houses, c_wifis = map(int, input().split())
houses = []
for _ in range(n_houses):
    houses.append(int(input()))
houses.sort()

max_interval = houses[-1] - houses[0]
installed_wifi = 1
while True:
    current_idx = 0
    installed_wifi = 1
    
    # print('interval: ', max_interval)
    for idx in range(1, len(houses)):
        # print(idx, f": {houses[current_idx]} + {max_interval} <= {houses[idx]}?")
        if houses[current_idx] + max_interval <= houses[idx]:
            current_idx = idx
            installed_wifi += 1
            # print(f"installed on {idx} house (total: {installed_wifi})")
    
    if installed_wifi < c_wifis:        
        max_interval -= 1
    else:
        break
    
print(max_interval)
```
